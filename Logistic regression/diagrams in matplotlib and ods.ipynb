## SEABORN

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in tips dataset
tips = sns.load_dataset('tips')
print(tips)

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# 1. Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()

# 2. Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=tips, x='size', y='total_bill', estimator='mean')
plt.title('Line Plot: Mean Total Bill by Party Size')
plt.show()

# 3. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Box Plot: Total Bill by Day')
plt.show()

# 4. Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x='day', y='total_bill')
plt.title('Violin Plot: Total Bill by Day')
plt.show()

# 5. Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title('Histogram: Distribution of Total Bill')
plt.show()

# 6. KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=tips['total_bill'], fill=True)
plt.title('KDE Plot: Total Bill Density')
plt.show()

# 7. Heatmap
plt.figure(figsize=(10, 6))
pivot_table = tips.pivot_table(values='total_bill', index='day', columns='time', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title('Heatmap: Mean Total Bill by Day and Time')
plt.show()

# 8. Pair Plot
sns.pairplot(tips, hue='sex')
plt.title('Pair Plot: Tips Dataset')
plt.show()

# 9. Facet Grid
g = sns.FacetGrid(tips, col='time', row='sex', margin_titles=True)
g.map(sns.histplot, 'total_bill')
plt.show()

# 10. Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='total_bill', estimator='mean')
plt.title('Bar Plot: Mean Total Bill by Day')
plt.show()

# 11. Count Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=tips, x='day')
plt.title('Count Plot: Number of Orders by Day')
plt.show()

# 12. Strip Plot
plt.figure(figsize=(10, 6))
sns.stripplot(data=tips, x='day', y='total_bill', jitter=True)
plt.title('Strip Plot: Total Bill by Day')
plt.show()

# 13. Swarm Plot
plt.figure(figsize=(10, 6))
sns.swarmplot(data=tips, x='day', y='total_bill')
plt.title('Swarm Plot: Total Bill by Day')
plt.show()

# 14. Joint Plot
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')
plt.title('Joint Plot: Total Bill vs Tip')
plt.show()

# 15. Residual Plot
plt.figure(figsize=(10, 6))
sns.residplot(data=tips, x='total_bill', y='tip', lowess=True)
plt.title('Residual Plot: Tips vs Total Bill')
plt.show()
```
